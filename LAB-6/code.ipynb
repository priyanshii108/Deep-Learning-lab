```python
# 1. Import Libraries
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Sample Sentence
sentence = ["I", "love", "deep", "learning"]

# 3. Assign Attention Weights (raw scores)
attention_scores = torch.tensor([0.1, 0.3, 0.4, 0.2])

# 4. Normalize using Softmax
attention_weights = F.softmax(attention_scores, dim=0).detach().numpy()

# 5. Print Weights (for clarity)
for word, weight in zip(sentence, attention_weights):
    print(f"{word}: {weight:.3f}")

# 6. Plot Heatmap
plt.figure(figsize=(8, 2))
sns.heatmap(
    [attention_weights],
    annot=True,
    fmt=".2f",
    xticklabels=sentence,
    yticklabels=["Attention"],
    cmap="Blues"
)

plt.title("Attention Heatmap Visualization")
plt.xlabel("Words")
plt.show()
```
