```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleGCN(nn.Module):
    def __init__(self):
        super(SimpleGCN, self).__init__()
        self.fc = nn.Linear(10, 2)

    def forward(self, x, adj):
        x = torch.matmul(adj, x)  # message passing
        x = self.fc(x)
        return F.relu(x)

# Dummy data
x = torch.rand((5, 10))          # 5 nodes, 10 features
adj = torch.eye(5)              # simple identity adjacency

model = SimpleGCN()
output = model(x, adj)

print("GCN Output:\n", output)
```
