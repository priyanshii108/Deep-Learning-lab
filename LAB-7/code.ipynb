```python
# 1. Install (run once if needed)
# pip install transformers tensorflow

# 2. Import Libraries
from transformers import BertTokenizer, TFBertForSequenceClassification
import tensorflow as tf
import numpy as np

# 3. Load Tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 4. Sample Data
texts = ["I love AI", "I hate bugs", "This is amazing", "This is terrible"]
labels = [1, 0, 1, 0]

# 5. Tokenization
encodings = tokenizer(
    texts,
    padding=True,
    truncation=True,
    max_length=32,
    return_tensors="tf"
)

labels = tf.convert_to_tensor(labels)

# 6. Load Pre-trained BERT Model
model = TFBertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=2
)

# 7. Compile Model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5),
    loss=model.compute_loss,
    metrics=['accuracy']
)

# 8. Train Model
model.fit(
    {'input_ids': encodings['input_ids'],
     'attention_mask': encodings['attention_mask']},
    labels,
    epochs=2,
    batch_size=2
)

# 9. Test Prediction
test_text = ["I really love this"]
test_enc = tokenizer(test_text, return_tensors="tf", padding=True, truncation=True)

pred = model.predict({
    'input_ids': test_enc['input_ids'],
    'attention_mask': test_enc['attention_mask']
})

predicted_class = np.argmax(pred.logits, axis=1)

print("\nTest Sentence:", test_text[0])
print("Predicted Sentiment:", "Positive" if predicted_class[0] == 1 else "Negative")
```
